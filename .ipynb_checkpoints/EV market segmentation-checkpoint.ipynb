```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings("ignore", category=UserWarning, module="sklearn")
import os
os.environ["OMP_NUM_THREADS"] = "1"
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
# Load the dataset
df = pd.read_csv("data.csv")
df.head()
```

```python
# Exploratory Data Analysis (EDA)
# Checking the shape of the dataset
print('Dataset Shape: ', df.shape)
```

```python
# Checking the info (columns, datatypes, nulls) of the dataset
print(' <<< DATASET INFORMATION -----------------------------------------------------------')
print(df.info())
```

```python
# Getting a statistical summary of the dataset
d = df.describe()
display('<<< DATASET STATISTICAL SUMMARY >>>', d)
```

```python
# Data visualization for EV range by brand
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Brand', y='Range_Km', palette='viridis', errorbar=None)
plt.xticks(rotation=45, ha='right')
plt.ylabel('Range (Km)', fontsize=14, family='serif')
plt.xlabel('Brand', family='serif', fontsize=14, labelpad=10)
plt.title(label='Electric Vehicle Range by Brand', weight=200, family='serif', size=15, pad=12)
plt.show()
```

```python
# Select features for PCA and KMeans
X = df[['Range_Km', 'Efficiency_WhKm', 'FastCharge_KmH', 'PriceEuro']]
```

```python
# Standardizing the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[['Range_Km', 'Efficiency_WhKm', 'FastCharge_KmH', 'PriceEuro']])
```

```python
# Applying PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
# Import necessary libraries
from sklearn.cluster import KMeans

# KMeans Clustering
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42, n_init=10)
    kmeans.fit(X_pca)
    wcss.append(kmeans.inertia_)
```

```python
# Plotting the Elbow Method
plt.figure(figsize=(8,6))
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method', fontsize=15)
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```

```python
# Selecting the optimal number of clusters (e.g., 3)
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42, n_init=10)
y_kmeans = kmeans.fit_predict(X_pca)

# Adding cluster labels to the original dataframe
df['Cluster'] = y_kmeans
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=y_kmeans, palette='viridis', s=100, alpha=0.8)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', marker='X')
plt.title('Clusters of Electric Vehicles', fontsize=15)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Range_Km'], bins=30, kde=True, color='blue')
plt.title('Distribution of Electric Vehicles by Range (Km)', fontsize=15)
plt.xlabel('Range (Km)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Cluster', y='PriceEuro', data=df, palette='viridis')
plt.title('Boxplot of Prices by Clusters', fontsize=15)
plt.xlabel('Cluster')
plt.ylabel('Price (Euro)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Range_Km', y='Efficiency_WhKm', hue='Cluster', data=df, palette='viridis', s=100, alpha=0.8)
plt.title('Range vs Efficiency by Clusters', fontsize=15)
plt.xlabel('Range (Km)')
plt.ylabel('Efficiency (Wh/Km)')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='FastCharge_KmH', data=df, palette='viridis')
plt.title('Count of Electric Vehicles by Fast Charge (Km/H)', fontsize=15)
plt.xlabel('Fast Charge (Km/H)')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```
